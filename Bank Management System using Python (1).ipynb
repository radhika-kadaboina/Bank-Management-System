# Bank Management System using Python

## Project Purpose:

##### This project is a **Bank Management System** developed using **Python and Tkinter**.  
##### The main aim of this project is to simulate basic banking operations using a **Graphical User Interface (GUI)**.

### Using this system, a user can:
- Create an account  
- Login  
- Deposit money  
- Withdraw money  
- View transaction history  
- Logout  

### This project is mainly created for **learning purposes**:
- To understand **Tkinter GUI programming**  
- To practice **Object-Oriented Programming (OOP)**  
- To learn **data storage using dictionary and list**


### Technologies Used

- **Python** – used for program logic  
- **Tkinter** – used to create GUI (windows, buttons, labels, etc.)  
- **OOP (Class & Object)** – used to structure the program  
- **Dictionary & List** – used to store user data and transaction history


### Main Class
+ The BankSystem class represents the complete banking system.
+ All functionalities such as account creation, login, deposit, withdraw, and transaction history are implemented inside this class.

## How the Project Works (Step by Step)

### Step 1: Project Initialization

+ First, a main window is created using Tkinter.This window acts as the base of the application where all other components are placed.
+  The project is designed using Object-Oriented Programming (OOP) by creating a main class called BankSystem.
+ This class controls the complete working of the application.

### Step 2: Data Structure for Storage

+ A dictionary is used to store all users’ details.
##### Each user’s data includes:

+ Name
+ Age
+ Salary
+ Balance
+ Transaction history
##### A separate variable is used to store the details of the currently logged-in user.
##### This approach allows handling multiple users in a single program execution.

### Step 3: Account Creation Process

##### A screen is provided where the user enters:
+ Name
+ Age
+ Salary
+ PIN
##### The system checks:
+ All fields are filled
+ PIN is unique
##### If valid, a new account is created and stored in the dictionary with:
+ Balance set to zero
+ Empty transaction history
##### This step simulates the process of opening a bank account.

### Step 4: Login Process

##### A login screen is provided where the user enters:

+ Name
+ PIN
##### The system verifies:

+ PIN exists in the dictionary
+ Name matches the stored name
##### If correct, the system logs the user in and displays the user dashboard.
##### This step represents authentication in a banking system.

### Step 5: User Dashboard

##### After login, a user screen is displayed showing:

+ User details (name, age, salary)
+ Current balance
##### The user is provided with options:
+ Deposit
+ Withdraw
+ View transaction history
+ Logout
##### This acts as the main control panel for the user.

### Step 6: Deposit Operation

##### When the user chooses deposit:

+ The system asks for an amount
+ The amount is added to the current balance
+ A transaction entry is created and stored in transaction history
##### This step simulates adding money to a bank account.

### Step 7: Withdraw Operation

##### When the user chooses withdraw:

+ The system asks for an amount
+ It checks if sufficient balance is available
+ If yes, the amount is deducted
+ A transaction entry is saved in transaction history
##### This step simulates withdrawing money from a bank account.

### Step 8: Transaction History

##### When the user selects transaction history:

+ A new window is opened
+ All past deposit and withdraw operations are displayed
+ Each transaction shows the type and updated balance
##### This helps the user track previous activities.

### Step 9: Logout Process

##### When the user logs out:

+ The user screen is hidden
+ The login screen is shown again
+ Current user data is cleared
##### This allows another user to log in without closing the program.

### Step 10: Data Handling Method

##### The system does not use any database.
##### All data is stored in memory using:

+ Dictionary for user details
+ List for transaction history

##### Because of this:

+ Data is temporary
+ When the program is closed, all data is lost

##### This approach is suitable for learning but not for real banking systems.

### Step 11: Additional Logic Using Person Class

##### To explain banking logic separately, a separate class named Person is created in another cell.

##### This class:

+ Stores name and balance
+ Handles deposit and withdraw operations
+ Maintains transaction history

##### This is used to demonstrate:

+ Core banking logic without GUI
+ Separation of business logic from user interface
+ Object-oriented design clearly

### Step 12: Overall Working Flow

+ User creates an account
+ User logs in
+ User performs deposit or withdraw
+ System updates balance
+ System records transaction
+ User views history
+ User logs out
##### This sequence represents the complete working of the project.


```python
from tkinter import *
from tkinter import messagebox, simpledialog

class BankSystem:
    def __init__(self, master):# init used to initialize the object 
        self.master = master
        self.master.title("Bank Management System")
        self.master.geometry("400x350")

        self.users = {}   # all users data
        self.current_user_data = None

        # ---------- Create Account Frame ----------
        self.create_account_frame = Frame(master)
        self.create_account_frame.pack(pady=10)

        Label(self.create_account_frame, text="Name").grid(row=0, column=0)
        Label(self.create_account_frame, text="Age").grid(row=1, column=0)
        Label(self.create_account_frame, text="Salary").grid(row=2, column=0)
        Label(self.create_account_frame, text="PIN").grid(row=3, column=0)

        self.name_entry = Entry(self.create_account_frame)
        self.age_entry = Entry(self.create_account_frame)
        self.salary_entry = Entry(self.create_account_frame)
        self.pin_entry = Entry(self.create_account_frame, show="*")

        self.name_entry.grid(row=0, column=1)
        self.age_entry.grid(row=1, column=1)
        self.salary_entry.grid(row=2, column=1)
        self.pin_entry.grid(row=3, column=1)

        Button(self.create_account_frame, text="Create Account", command=self.create_account).grid(row=4, column=1)

        # ---------- Login Frame ----------
        self.login_frame = Frame(master)
        self.login_frame.pack(pady=10)

        Label(self.login_frame, text="Name").grid(row=0, column=0)
        Label(self.login_frame, text="PIN").grid(row=1, column=0)

        self.login_name_entry = Entry(self.login_frame)
        self.login_pin_entry = Entry(self.login_frame, show="*")

        self.login_name_entry.grid(row=0, column=1)
        self.login_pin_entry.grid(row=1, column=1)

        Button(self.login_frame, text="Login", command=self.login).grid(row=2, column=1)

        # ---------- User Frame ----------
        self.user_frame = Frame(master)

        self.info_label = Label(self.user_frame, text="")
        self.info_label.pack()

        self.balance_label = Label(self.user_frame, text="Balance: 0")
        self.balance_label.pack()

        Button(self.user_frame, text="Deposit", command=self.deposit).pack(pady=2)
        Button(self.user_frame, text="Withdraw", command=self.withdraw).pack(pady=2)
        Button(self.user_frame, text="View Transaction History", command=self.view_transaction_log).pack(pady=2)
        Button(self.user_frame, text="Logout", command=self.logout).pack(pady=2)

    # ---------- Create Account ----------
    def create_account(self):
        name = self.name_entry.get()
        age = self.age_entry.get()
        salary = self.salary_entry.get()
        pin = self.pin_entry.get()

        if not name or not age or not salary or not pin:
            messagebox.showerror("Error", "All fields required")
            return

        if pin in self.users:
            messagebox.showerror("Error", "PIN already exists")
            return

        self.users[pin] = {
            "name": name,
            "age": age,
            "salary": salary,
            "balance": 0,
            "transaction_log": []
        }

        messagebox.showinfo("Success", "Account Created")

        self.name_entry.delete(0, END)
        self.age_entry.delete(0, END)
        self.salary_entry.delete(0, END)
        self.pin_entry.delete(0, END)

    # ---------- Login ----------
    def login(self):
        name = self.login_name_entry.get()
        pin = self.login_pin_entry.get()

        if pin in self.users and self.users[pin]["name"] == name:
            self.current_user_data = self.users[pin]

            self.info_label.config(text=f"Name: {self.current_user_data['name']}  Age: {self.current_user_data['age']}  Salary: {self.current_user_data['salary']}")
            self.balance_label.config(text=f"Balance: {self.current_user_data['balance']}")

            self.create_account_frame.pack_forget()
            self.login_frame.pack_forget()
            self.user_frame.pack(pady=10)
        else:
            messagebox.showerror("Error", "Invalid Login")

    # ---------- Deposit ----------
    def deposit(self):
        amount = simpledialog.askstring("Deposit", "Enter amount:")

        if not amount or not amount.isdigit():
            return

        amount = int(amount)
        self.current_user_data["balance"] += amount

        self.current_user_data["transaction_log"].append(
            f"Deposit +{amount} | Balance {self.current_user_data['balance']}"
        )

        self.balance_label.config(text=f"Balance: {self.current_user_data['balance']}")

    # ---------- Withdraw ----------
    def withdraw(self):
        amount = simpledialog.askstring("Withdraw", "Enter amount:")

        if not amount or not amount.isdigit():
            return

        amount = int(amount)

        if amount > self.current_user_data["balance"]:
            messagebox.showerror("Error", "Insufficient Balance")
            return

        self.current_user_data["balance"] -= amount

        self.current_user_data["transaction_log"].append(
            f"Withdraw -{amount} | Balance {self.current_user_data['balance']}"
        )

        self.balance_label.config(text=f"Balance: {self.current_user_data['balance']}")

    # ---------- View History ----------
    def view_transaction_log(self):
        win = Toplevel(self.master)
        win.title("Transaction History")

        lb = Listbox(win, width=50)
        lb.pack(padx=10, pady=10)

        for t in self.current_user_data["transaction_log"]:
            lb.insert(END, t)

    # ---------- Logout ----------
    def logout(self):
        self.user_frame.pack_forget()
        self.login_frame.pack(pady=10)

        self.login_name_entry.delete(0, END)
        self.login_pin_entry.delete(0, END)

        self.current_user_data = None


def main():
    root = Tk()
    app = BankSystem(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```

# dynamic persons details

```python
class Person:
    def __init__(self, name, balance=0):
        self.name = name
        self.balance = balance
        self.history = []

    def deposit(self, amount):
        self.balance += amount
        self.history.append(f"Deposited {amount}")
        print("Deposit successful")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            self.history.append(f"Withdraw {amount}")
            print("Withdraw successful")
        else:
            print("Insufficient balance")

    def show_history(self):
        print("Transaction History:")
        for h in self.history:
            print(h)
```
